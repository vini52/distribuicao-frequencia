#DISTRIBUIÇÃO DE FREQUENCIA

```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv('/content/drive/MyDrive/dados_tratados.csv', sep=",", encoding='iso-8859-1')
```

```python
df.shape
```

```python
compra = df['compra']
```

```python
compra
```

##TABELA DE FREQUENCIAS

```python
# Frequencia absoluta
freq_absoluta = Counter(compra)
freq_absoluta
```

```python
# Transformando em dataframe
tabela = pd.DataFrame.from_dict(freq_absoluta, orient='index')
tabela
```

```python
#Colocar em ordem crescente
tabela = tabela.sort_index(ascending=True)
tabela
```

```python
#Colocar os indices reais
tabela.reset_index(inplace=True)
tabela
```

```python
#Renomear colunas
tabela = tabela.rename(columns={'index': 'valor'})
tabela.head()
```

```python
tabela = tabela.rename(columns={0: 'freq_absoluta'})
tabela.head()
```

```python
tabela['freq_rel'] = tabela['freq_absoluta'] / tabela['freq_absoluta'].sum()
tabela
```

```python
#Incluir frequencia ercentual relativa
tabela['freq_rel_perc'] = tabela['freq_rel'] * 100
tabela
```

```python
#Incluir frequencia acumulada
tabela['freq_acum'] = tabela['freq_absoluta'].cumsum()
tabela
```

##TABELA DE CLASSE DE FREQUENCIA

```python
tabela.valor.min()
```

```python
tabela.valor.max()
```

```python
dif = tabela.valor.max() - tabela.valor.min()
dif
```

```python
classes = [4.7, 4.8, 4.9, 5, 5.1, 5.2, 5.3]
```

```python
labels = ['4.7 - 4.8', '4.8 - 4.9', '4.9 - 5', '5 - 5.1', '5.1 - 5.2', '5.2 - 5.3']
```

```python
intervalos = pd.cut(x = tabela.valor, bins = classes, labels = labels, include_lowest = True)
intervalos
```

```python
freq_absoluta = pd.value_counts(intervalos)
freq_absoluta
```

```python
freq_rel = pd.value_counts(intervalos, normalize=True)
freq_rel
```

```python
dist_freq = pd.DataFrame({'Frequencia absoluta': freq_absoluta,
                          'Frequencia relativa': freq_rel})
dist_freq
```

```python
dist_freq.sort_index(ascending=True, inplace=True)
dist_freq
```

```python
dist_freq['freq_rel_perc'] = np.round(dist_freq['Frequencia relativa'] * 100, 2)
dist_freq
```

```python
dist_freq['freq_acum'] = dist_freq['Frequencia absoluta'].cumsum()
dist_freq
```
